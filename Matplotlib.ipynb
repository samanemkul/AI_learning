```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)

plt.title("First line plot")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='b',linestyle=':',linewidth=2)
plt.show()
```

```python
x
```

```python
y1=2*x
y2=3*x
```

```python
plt.plot(x,y1,color='g',linewidth=2)
plt.plot(x,y2,color='b',linewidth=3)
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='g',linewidth=2)
plt.subplot(1,2,2)
plt.plot(x,y2,color='b',linewidth=3)
plt.show()
#to plot two lines in two different graphs
```

```python
#Bar plot
student ={"Luniva":97,"Ivan":87,"Razisha":56}
names=list(student.keys())
values=list(student.values())
```

```python
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title("Marks of students")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
#horizontal barplot
```

```python
plt.barh(names,values)
plt.title("Marks of students")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[1,34,5,64,7,35,89,32,44]
plt.scatter(x,y)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,80,90]
y=[1,34,5,64,7,35,89,32,44]
a=[1,2,3,34,50,79,78,9,39]
plt.scatter(x,y,marker="*",c='g',s=100)
plt.scatter(x,a,marker=".",c='b',s=200)
plt.show()
```
